```python
# Titanic Data Analysis Project
# Author: Urvi Sharda

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic Dataset (comes from seaborn)
titanic = sns.load_dataset('titanic')

# Display first 5 rows
print("First 5 rows of the dataset:")
print(titanic.head())

# Basic info
print("\nBasic info about dataset:")
print(titanic.info())

# Basic stats
print("\nSummary statistics:")
print(titanic.describe())

# Count passengers by class
class_count = titanic['class'].value_counts()
print("\nPassenger count by class:")
print(class_count)

# Plot: Survival count
plt.figure(figsize=(6,4))
sns.countplot(x='survived', data=titanic, palette='Set2')
plt.title("Survival Count")
plt.savefig("survival_count.png")
plt.show()

# Plot: Class vs Survival
plt.figure(figsize=(6,4))
sns.countplot(x='class', hue='survived', data=titanic, palette='Set1')
plt.title("Class vs Survival")
plt.savefig("class_vs_survival.png")
plt.show()
```
